```python
def request_until_succeed(url):
    req = urllib2.Request(url)
    success = False
    is400 = False
    is100 = False
    while success is False and is400 is False and is100 is False:
        try:
            is400 = False
            is100 = False
            response = urllib2.urlopen(req)
            if response.getcode() == 200:
                success = True
        except Exception, e:
            print e
            time.sleep(5)
            
            print "Error for URL %s: %s" % (url, datetime.datetime.now())
            
            if "400" in str(e):
                print "400 error"
                is400 = True
                
            if "100" in str(e):
                print "100 error"
                is100 = True    

    if is400 is False and is100 is False:
        return response.read()
    else:
        return "{}"
    
```

```python
def getFacebookPageFeedData(page_id, access_token, num_statuses):
    
    # construct the URL string
    base = "https://graph.facebook.com/v2.8"
    node = "/" + page_id + "/feed" 
    parameters = "/?fields=message,link,created_time,type,name,id,likes.limit(1).summary(true),comments.limit(1).summary(true),shares&limit=%s&access_token=%s" % (num_statuses, access_token) # changed
    url = base + node + parameters
    
    #print url
    
    # retrieve data
    data = json.loads(request_until_succeed(url))
    
    return data
```

```python
def processFacebookPageComments(comments_dump):
    
    comments_list_only = [] 
    
    has_next_page = True
    num_processed = 0   # keep a count on how many we've processed
    scrape_starttime = datetime.datetime.now()
    new_comments = comments_dump
    if 'data' in new_comments.keys():
        for c in new_comments['data']:
            comments_list_only.append(c) 
            if c['comment_count'] > 0:
                #print(c['comment_count'])
                #print(c['id'])
                comments_list_only.extend(
                    processFacebookPageComments(
                        getFacebookCommentData(c['id'],access_token,100)))
        #print('--------------')
        #print(new_comments['data'])
        #comments_list_only.extend(new_comments['data']) 
    else:
        has_next_page = False
    
    while has_next_page:
        num_processed += 1
        #print "%s Pages Processed: %s" % (num_processed, datetime.datetime.now())
        
        #print (new_comments.keys())

        # if there is no next page, we're done.
        
        if 'paging' in new_comments.keys():
            if 'next' in new_comments['paging'].keys():
                new_comments = json.loads(request_until_succeed(new_comments['paging']['next']))
                for c in new_comments['data']:
                    comments_list_only.append(c)
                    if c['comment_count'] > 0:
                        #print(c['comment_count'])
                        #print(c['id'])
                        comments_list_only.extend(
                            processFacebookPageComments(
                                getFacebookCommentData(c['id'],access_token,100)))
                #print('--------------')
                #comments_list_only.extend(new_comments['data'])
            else:
                has_next_page = False                        

        else:
            has_next_page = False
    
    
    return comments_list_only
```

```python
def makeCommentsIntoRows(post_id,comments_array):
    
    # Additionally, some items may not always exist,
    # so must check for existence first
    
    #uni_writer.writerow(["post_id", "status_message", "link_name", "status_type", "status_link",
    #   "status_published", "num_likes", "num_comments", "num_shares", "likes"])
    
    #message,from,like_count,message_tags,parent,created_time,likes
    
    comments_in_rows = []
    
    for comment in comments_array:
    
        page_id = post_id.split('_')[0]
        comment_id = '' if 'id' not in comment.keys() else comment['id']
        # COMMENT MESSAGE NOT INCLUDED!!! Uncomment below:
        message = '' if 'message' not in comment.keys() else comment['message'].encode('utf-8')
        if 'from' in comment.keys():
            from_id = '' if 'id' not in comment['from'].keys() else comment['from']['id']
            from_name = '' if 'name' not in comment['from'].keys() else comment['from']['name'].encode('utf-8')
        else: 
            from_id = ''
            from_name = ''
        like_count = '' if 'like_count' not in comment.keys() else comment['like_count']
        comment_count = '' if 'comment_count' not in comment.keys() else comment['comment_count']
        
        message_tags = [] if 'message_tags' not in comment.keys() \
            else json.dumps(comment['message_tags'], indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
        parent = [] if 'parent' not in comment.keys() \
            else json.dumps(comment['parent'], indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
        likes = [] if 'likes' not in comment.keys() \
            else json.dumps(comment['likes']['data'], indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
        sub_comments = [] if 'comments' not in comment.keys() \
            else json.dumps(comment['comments']['data'], indent=4, sort_keys=True, ensure_ascii=False).encode('utf-8')
        
        comment_published = datetime.datetime.strptime(comment['created_time'],'%Y-%m-%dT%H:%M:%S+0000')
        comment_published = comment_published + datetime.timedelta(hours=-5) # EST
        comment_published = comment_published.strftime('%Y-%m-%d %H:%M:%S') # best time format for spreadsheet programs
        
        comments_in_rows.append((page_id,comment_id,message,from_id,from_name,comment_published,like_count,
            comment_count,message_tags,parent,likes,sub_comments))
        
    return comments_in_rows
```

```python
def scrapeFacebookPagePostsAndComments(page_id, access_token):
    
    with open('%s_facebook_posts_and_comments.csv' % page_id, 'wb') as file:
        #w = csv.writer(file)
        uni_writer = UnicodeWriter(file)
        #uni_writer.writerow(["status_id", "status_message", "link_name", "status_type", "status_link",
        #   "status_published", "num_posts_likes", "num_post_comments", "num_shares", "page_id", "comment_id", "message", "from_id", 
        #   "from_name", "comment_published", "num_comment_likes", "num_sub_comments", "message_tags", "parent",
        #   "likes", "sub_comments"])
        uni_writer.writerow(["status_id", "link_name", "status_type", "status_link",
           "status_published", "num_posts_likes", "num_post_comments", "num_shares", "page_id", "comment_id", "message", "from_id", 
           "from_name", "comment_published", "num_comment_likes", "num_sub_comments", "message_tags", "parent",
           "likes", "sub_comments"])

        has_next_page = True
        num_processed = 0   # keep a count on how many we've processed
        restrict_time = False
        scrape_starttime = datetime.datetime.now()
        
        print "Scraping %s Facebook Page: %s\n" % (page_id, scrape_starttime)
        
        statuses = getFacebookPageFeedData(page_id, access_token, 100)
        
        while has_next_page:
            for status in statuses['data']:
                
                row = processFacebookPageFeedStatus(status)
                #print row[5]
                #date_object = datetime.datetime.strptime(row[5], '%Y-%m-%d %H:%M:%S')
                date_object = datetime.datetime.strptime(row[4], '%Y-%m-%d %H:%M:%S')
                if restrict_time == True: 
                    if (date_object.month == 9 \
                    #and (date_object.year == 2016)):
                    and (date_object.year == 2016 or date_object.year == 2014 or date_object.year == 2012)):
                        comments_on_post = makeCommentsIntoRows(
                            status['id'],processFacebookPageComments(
                            getFacebookCommentData(status['id'], access_token,100)))

                        #print(row)
                        for comment_row in comments_on_post:
                            row_full = row + comment_row

                            uni_writer.writerow(row_full)
                            #w.writerow(row)

                            #print "%s \n" % (row) 

                            # output progress occasionally to make sure code is not stalling
                            num_processed += 1
                            if num_processed % 1000 == 0:
                                print "%s Comments Processed: %s" % (num_processed, datetime.datetime.now())
                else:
                    comments_on_post = makeCommentsIntoRows(
                        status['id'],processFacebookPageComments(
                        getFacebookCommentData(status['id'], access_token,100)))

                    #print(row)
                    for comment_row in comments_on_post:
                        row_full = row + comment_row

                        uni_writer.writerow(row_full)
                        #w.writerow(row)

                        #print "%s \n" % (row) 

                        # output progress occasionally to make sure code is not stalling
                        num_processed += 1
                        if num_processed % 1000 == 0:
                            print "%s Comments Processed: %s" % (num_processed, datetime.datetime.now())                    

            # if there is no next page, we're done.
            if 'paging' in statuses.keys():
                statuses = json.loads(request_until_succeed(statuses['paging']['next']))
            else:
                has_next_page = False
                
        
        print "\nDone!\n%s Comments Processed in %s" % (num_processed, datetime.datetime.now() - scrape_starttime)
```
